```
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img=cv2.imread('white_female.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=face_cascade.detectMultiScale(gray,1.3,5)
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),10)
  
plt.grid(None)
plt.xticks([])
plt.yticks([])
imgplot=plt.imshow(img)
print(x,y,x+w,y+h)
cropped_img=gray[y:y+h,x:x+w]
plt.imshow(img)
    
```

```
landmark_detector=cv2.face.createFacemarkLBF()
landmark_detector.loadModel("Landmark_detection.txt")
landmarks=landmark_detector.fit(gray,faces)
for landmark in landmarks[1][0]:
  for element in landmark:
    cv2.circle(cropped_img,(element[0],element[1]),1,(255,0,0),1)

plt.imshow(cropped_img)
```

```
def rect_to_bb(rect):
	# take a bounding predicted by dlib and convert it
	# to the format (x, y, w, h) as we would normally do
	# with OpenCV
	x = rect.left()
	y = rect.top()
	w = rect.right() - x
	h = rect.bottom() - y
	# return a tuple of (x, y, w, h)
	return (x, y, w, h)
```
